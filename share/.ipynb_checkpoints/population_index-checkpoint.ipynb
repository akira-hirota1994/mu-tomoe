```python
#　Observation time
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
from IPython.display import clear_output
from matplotlib.backends.backend_pdf import PdfPages

df_mu = pd.read_csv("MU_data_2009-2015.csv")
day = df_mu["DD"].values
hour = df_mu["hh"].values
minute = df_mu["mm"].values
time = day*24.*60. + hour*60. + minute

### total observation time [min]
total_observation_time = []
mjd = df_mu["MJD"].values
dif_mjd = mjd
for i in range(len(mjd)-1):
    dif_mjd[i] = mjd[i+1] - mjd[i]
separate_index = np.where(dif_mjd > 0.02) #about 30min
separate_index = list(separate_index)
separate_index = np.insert(separate_index, 0 , 0)

for j in range(len(separate_index)-1):
    observation_time = time[separate_index[j+1]] - time[separate_index[j]+1]
    total_observation_time = np.hstack((total_observation_time, observation_time))

total_observation_time = np.sum(total_observation_time)
total_observation_time_hour = total_observation_time/60.
print "observation time is",total_observation_time, "[min]"
print "observation time is",total_observation_time_hour, "[hour]"
print "Complete!"
```

```python
#　Cumulative number colliding with the Earth per year
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
from IPython.display import clear_output
from matplotlib.backends.backend_pdf import PdfPages

df_mu = pd.read_csv("MU_data_2009-2015.csv")
df_fireball = pd.read_csv("fireball.csv", header=None)
df_gruen = pd.read_csv("Gruen.csv", header=None)

array_mu = df_mu.values
array_fireball = df_fireball.values
array_gruen = df_gruen.values
array_gruen = 10**(array_gruen)

radial_distance = np.pi*(100*np.tan(np.deg2rad(4.)))**2
mag = df_mu["mass_grams_calculated"].values
mag2 = df_mu["size_calculated"].values
total_observation_time_hour = 845.8 #[hour]
radius_of_the_earth = 6378.1 #[km]
surface_area_of_the_earth = 4.*np.pi*radius_of_the_earth**2
year_hr = 365.*24.

def sizeindex(x, a, b):
    return 10**(b-a*np.log10(x))

def massindex(x, a, b):
    return b*x**(-1*(a-1))
    
#################################################################sizeごとにテーブル
array2_mu1 = array_mu[np.where((mag2>4.57078150e-05) & (mag2<9.84745022e-05))]
array2_mu2 = array_mu[np.where((mag2>9.84745022e-05) & (mag2<2.12156884e-04))]
array2_mu3 = array_mu[np.where((mag2>2.12156884e-04) & (mag2<4.57078150e-04))]
array2_mu4 = array_mu[np.where((mag2>4.57078150e-04) & (mag2<9.84745022e-04))]
array2_mu5 = array_mu[np.where((mag2>9.84745022e-04) & (mag2<2.12156884e-03))]
array2_mu6 = array_mu[np.where((mag2>2.12156884e-03) & (mag2<4.57078150e-03))]
array2_mu7 = array_mu[np.where((mag2>4.57078150e-03) & (mag2<9.84745022e-03))]
array2_mu8 = array_mu[np.where((mag2>9.84745022e-03) & (mag2<2.12156884e-02))]
array2_mu9 = array_mu[np.where((mag2>2.12156884e-02) & (mag2<4.57078150e-02))]

radial2_distance1 = np.pi*np.average(array2_mu1[:,-2])**2
radial2_distance2 = np.pi*np.average(array2_mu2[:,-2])**2
radial2_distance3 = np.pi*np.average(array2_mu3[:,-2])**2
radial2_distance4 = np.pi*np.average(array2_mu4[:,-2])**2
radial2_distance5 = np.pi*np.average(array2_mu5[:,-2])**2
radial2_distance6 = np.pi*np.average(array2_mu6[:,-2])**2
radial2_distance7 = np.pi*np.average(array2_mu7[:,-2])**2
radial2_distance8 = np.pi*np.average(array2_mu8[:,-2])**2
radial2_distance9 = np.pi*np.average(array2_mu9[:,-2])**2

num2_sum1 = np.sum(array2_mu1, axis=0)[-1]/total_observation_time_hour/radial2_distance1*\
                surface_area_of_the_earth*year_hr
num2_sum2 = np.sum(array2_mu2, axis=0)[-1]/total_observation_time_hour/radial2_distance2*\
                surface_area_of_the_earth*year_hr
num2_sum3 = np.sum(array2_mu3, axis=0)[-1]/total_observation_time_hour/radial2_distance3*\
                surface_area_of_the_earth*year_hr
num2_sum4 = np.sum(array2_mu4, axis=0)[-1]/total_observation_time_hour/radial2_distance4*\
                surface_area_of_the_earth*year_hr
num2_sum5 = np.sum(array2_mu5, axis=0)[-1]/total_observation_time_hour/radial2_distance5*\
                surface_area_of_the_earth*year_hr
num2_sum6 = np.sum(array2_mu6, axis=0)[-1]/total_observation_time_hour/radial2_distance6*\
                surface_area_of_the_earth*year_hr
num2_sum7 = np.sum(array2_mu7, axis=0)[-1]/total_observation_time_hour/radial2_distance7*\
                surface_area_of_the_earth*year_hr
num2_sum8 = np.sum(array2_mu8, axis=0)[-1]/total_observation_time_hour/radial2_distance8*\
                surface_area_of_the_earth*year_hr
num2_sum9 = np.sum(array2_mu9, axis=0)[-1]/total_observation_time_hour/radial2_distance9*\
                surface_area_of_the_earth*year_hr

num2_sum1_err = (np.sqrt(np.sum(array2_mu1, axis=0)[-1]))/total_observation_time_hour/radial2_distance1*\
                surface_area_of_the_earth*year_hr
num2_sum2_err = (np.sqrt(np.sum(array2_mu2, axis=0)[-1]))/total_observation_time_hour/radial2_distance2*\
                surface_area_of_the_earth*year_hr
num2_sum3_err = (np.sqrt(np.sum(array2_mu3, axis=0)[-1]))/total_observation_time_hour/radial2_distance3*\
                surface_area_of_the_earth*year_hr
num2_sum4_err = (np.sqrt(np.sum(array2_mu4, axis=0)[-1]))/total_observation_time_hour/radial2_distance4*\
                surface_area_of_the_earth*year_hr
num2_sum5_err = (np.sqrt(np.sum(array2_mu5, axis=0)[-1]))/total_observation_time_hour/radial2_distance5*\
                surface_area_of_the_earth*year_hr
num2_sum6_err = (np.sqrt(np.sum(array2_mu6, axis=0)[-1]))/total_observation_time_hour/radial2_distance6*\
                surface_area_of_the_earth*year_hr
num2_sum7_err = (np.sqrt(np.sum(array2_mu7, axis=0)[-1]))/total_observation_time_hour/radial2_distance7*\
                surface_area_of_the_earth*year_hr
num2_sum8_err = (np.sqrt(np.sum(array2_mu8, axis=0)[-1]))/total_observation_time_hour/radial2_distance8*\
                surface_area_of_the_earth*year_hr
num2_sum9_err = (np.sqrt(np.sum(array2_mu9, axis=0)[-1]))/total_observation_time_hour/radial2_distance9*\
                surface_area_of_the_earth*year_hr


b2= np.hstack((num2_sum1,num2_sum2,num2_sum3,num2_sum4,num2_sum5,num2_sum6,num2_sum7,num2_sum8,num2_sum9))
b2_err = np.hstack((num2_sum1_err,num2_sum2_err,num2_sum3_err,num2_sum4_err,num2_sum5_err,num2_sum6_err,\
                    num2_sum7_err,num2_sum8_err,num2_sum9_err))

b2 = b2[::-1]
b2_err = b2_err[::-1]
b2 = np.cumsum(b2)
b2_err = np.cumsum(b2_err)
a2 = np.array([9.84745022e-05, 2.12156884e-05, 4.57078150e-04, 9.84745022e-04,\
               2.12156884e-03, 4.57078150e-03, 9.84745022e-03, 2.12156884e-02, 4.57078150e-02])
a2 = a2[::-1]
clear_output()

##################################################################

param2, cov2 = curve_fit(sizeindex, a2[0:5], b2[0:5])
y_fit2 = 10**(param2[1]-param2[0]*np.log10(a2))
a3 = np.logspace(-1.5,9,10)
p_brown = 10**(1.568-2.7*np.log10(a3))

array_fireball_x = array_fireball[:,0]
array_fireball_y = 10**array_fireball[:,1]
array_gruen_x = array_gruen[:,0]
array_gruen_y = array_gruen[:,1]

fig = plt.figure(figsize=(10,10))
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams["font.size"] = 25
ax1 = fig.add_subplot(111)
ax1.plot(a2,y_fit2,linewidth=4,linestyle="dashdot",color="red",label="This research")
ax1.plot(a3,p_brown,linewidth=4,linestyle="dashdot",color="green",label="P.Brown(2002)")
ax1.errorbar(a2,b2,yerr=b2_err,fmt='ro',ecolor='red',mfc="red",capsize=4)
ax1.plot(array_fireball_x,array_fireball_y,color="b",linestyle="solid",linewidth=5,label="Halliday(1996)")
ax1.plot(array_gruen_x,array_gruen_y,color="purple",linestyle="dashdot",label="Gruen(1985)")
ax1.legend(loc='upoer right')
ax1.set_ylabel("Log Cumulative number colliding with the Earth per year")
ax1.set_xlabel("Log Particle Diameter[m]")
ax1.set_xscale("log")
ax1.grid(which='major',alpha=0.2,color='black',linestyle='-')
ax1.set_xlim(10**-8,10**3)
plt.yscale("log")
plt.xscale("log")
plt.ylim(10e-7,10e20)
plt.savefig("size_index.png")
plt.show()

clear_output()
print "Complete!"
```
