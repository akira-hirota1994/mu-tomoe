```python
###   Create script file
import pandas as pd
import numpy as np
import os
import subprocess
#*** need to creat ~.sh file before run this program.
################################ xpase - ds9で一枚一枚画像開いてカタログファイル(.xml)を出させる
df = pd.read_csv("simultaneous_event_candidates_table_MU+Tomo-e6.csv")
df["ID_fits"] = "d_fits/sub" + df["ID"] + ".fits"
df["ID_xml"] = "sub" + df["ID"] + ".xml"
df["ID_xml_gaia"] = "sub" + df["ID"] + "_gaia.xml"
df["ID_pdf"] = "sub" + df["ID"] + "_lightcurve.pdf" #流星のphotometry
df["ID_csv"] = "sub" + df["ID"] + "_star_used.csv" #csv

ID_list_fits = df["ID_fits"].values
ID_list_xml = df["ID_xml"].values
ID_list_xml_gaia = df["ID_xml_gaia"].values
ID_list_xml_ucac = "sub" + df["ID"] + "_ucac.xml"
ID_list_pdf = df["ID_pdf"].values
ID_list_csv = df["ID_csv"].values
ID_list_phot_gaia = "sub" + df["ID"] + "_ucac.pdf"
##################################################################
##################################################  ds9でカタログ恒星取得
match_catalog_tycho = "xpaset -p ds9 catalog cds Tycho-2\n" # カタログ _ Tycho-2
match_catalog_gaia = "xpaset -p ds9 catalog cds {I/345}\n" # カタログ _ Gaia
match_catalog_ucac = "xpaset -p ds9 catalog cds {I/322A}\n" # カタログ _ USNO UCAC4
clear_catalog = "xpaset -p ds9 catalog clear\n"
close_catalog = "xpaset -p ds9 catalog close\n"
with open("ds9_catalog_tycho.sh", mode='w') as f:
    f.write("#!/bin/bash\n")
    for (i, j) in zip(ID_list_fits, ID_list_xml):
        open_fits = "xpaset ds9 fits " + str(i) + " < "  + str(i) + "\n"
        save_catalog = "xpaset -p ds9 catalog save " + str(j) + "\n"
        f.write(open_fits)
        f.write(match_catalog_tycho)
        f.write(save_catalog)
        f.write(clear_catalog)
        f.write(close_catalog)
# こっちはGaia 
with open("ds9_catalog_gaia.sh", mode='w') as f:
    f.write("#!/bin/bash\n")
    for (i, j) in zip(ID_list_fits, ID_list_xml_gaia):
        open_fits = "xpaset ds9 fits " + str(i) + " < "  + str(i) + "\n"
        save_catalog = "xpaset -p ds9 catalog save " + str(j) + "\n"
        f.write(open_fits)
        f.write(match_catalog_gaia)
        f.write(save_catalog)
        f.write(clear_catalog)
        f.write(close_catalog)
with open("ds9_catalog_ucac.sh", mode='w') as f:
    f.write("#!/bin/bash\n")
    for (i, j) in zip(ID_list_fits, ID_list_xml_ucac):
        open_fits = "xpaset ds9 fits " + str(i) + " < "  + str(i) + "\n"
        save_catalog = "xpaset -p ds9 catalog save " + str(j) + "\n"
        f.write(open_fits)
        f.write(match_catalog_ucac)
        f.write(save_catalog)
        f.write(clear_catalog)
        f.write(close_catalog)
##################################################################
###########################################  出力した画像諸々を各ディレクトリに移動
with open("mv_file.sh", mode='w') as f:
    f.write("#!/bin/bash\n")
    for (i, j, k, l, m, n) in zip(ID_list_xml, ID_list_pdf, ID_list_xml_gaia, ID_list_csv,\
                                    ID_list_phot_gaia, ID_list_xml_ucac):
        #mv_xml = "mv " + str(i) + " d_xml_tycho/ 2>/dev/null\n"
        #mv_xml_gaia = "mv " + str(k) + " d_xml_gaia/ 2>/dev/null\n"
        #mv_xml_ucac = "mv " + str(q) + " d_xml_ucac/ 2>/dev/null\n"
        mv_pdf = "mv " + str(j) + " d_lightcurve/ 2>/dev/null\n"
        #mv_csv = "mv " + str(o) + " d_csv_ucac_used2/ 2>/dev/null\n"
        #mv_phot_gaia = "mv " + str(p) + " d_photometry_ucac/ 2>/dev/null\n"
        #f.write(mv_xml)
        #f.write(mv_xml_gaia)
        #f.write(mv_xml_ucac)
        #f.write(mv_png)
        #f.write(mv_phot)
        #f.write(mv_rot)
        f.write(mv_pdf)
        #f.write(mv_csv)
        #f.write(mv_phot_gaia)
##################################################################
###########################################  Tomo-e　のfitsファイルを移動
file_name = os.listdir("./d_fits/")
fits_file = [s for s in file_name if 's' in s]
with open("mv_fits.sh", mode='w') as f:
    f.write("#!/bin/bash\n")
    for i in fits_file:
        mv_fits = "mv " + str(i) + " d_fits/ 2>/dev/null\n"
        f.write(mv_fits)
##################################################################
############################################################# Github        
with open("git.sh", mode='w') as f:
    f.write("#!/bin/bash\n")
    f.write("git add mu-tomoe_analysis.ipynb\n")#add
    f.write("git add mu-tomoe_analysis2.ipynb\n")#add
    f.write("git add mu-tomoe_analysis3.ipynb\n")#add
    f.write("git add photometry_meteor.m\n")#add
    f.write("git add light_curve.m\n")#add
    f.write("git add photometry_ucac.ipynb\n")#add
    f.write("git add photometry_gaia.ipynb\n")#add
    f.write("git add script.ipynb\n")#add
    f.write("git add recheck_photometry.csv\n")#add
    f.write("git add xml_gaia.ipynb\n")#add
    f.write("git add xml_ucac.ipynb\n")#add
    f.write("git add gband2vband.ipynb\n")#add
    f.write("git add gband2vband.pdf\n")#add
    f.write("git add simultaneous_event_candidates_table_MU+Tomo-e10.csv\n")
    f.write("git add simultaneous_event_candidates_table_MU+Tomo-e10_full.csv\n")
    f.write("git add d_dcriterion\n")
    f.write("git add d_lightcurve\n")
    f.write("git add share\n")
    f.write("git add d_image\n")#add
    f.write("git commit -m \"akira.h\"\n")#commit
    f.write("git push origin master\n")#push

#subprocess.call(["mv ds9_catalog.sh d_shellscript"], shell = True)
#subprocess.call(["mv mv_imagefile.sh d_shellscript"], shell = True)
#subprocess.call(["mv mv_fits.sh d_shellscript"], shell = True)
print "Complete!" 
```
